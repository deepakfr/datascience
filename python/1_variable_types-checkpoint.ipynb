## Type conversion (casting)

```python
my_num = 7
print(my_num)
print(type(my_num))
```

```python
# integer to string
str(my_num)
```

```python
type(str(my_num))
```

```python
my_num
```

```python
my_num_string = str(my_num)
print(my_num_string)
print(type(my_num_string))
```

```python
# string to integer
my_num_int = int(my_num_string)
print(my_num_int)
print(type(my_num_int))
```

```python
# integer to float
my_num_float = float(my_num_int)
print(my_num_float)
print(type(my_num_float))
```

```python
# float to boolean
my_num_bool = bool(my_num_float)
print(my_num_bool)
print(type(my_num_bool))
```

## String operations

```python
my_word = 'tables'
```

```python
#We want to acess the 3rd letter of our word
my_word[2] #strings are indexed from zero
```

```python
my_word[-4] #negative indexing
```

```python
my_word[1:3]  #slicing with 2 limits = defining a substring
```

```python
my_word[2:]  #slicing with 1 limit at the beginning
```

```python
my_word[:4]  #slicing with 1 limit at the end
```

```python
my_word[1:-1:2] #slicing with 2 limits and a step
```

```python
my_word[::-1] #inversion of our string
```
