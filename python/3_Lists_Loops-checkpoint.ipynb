## Loops

```python
numbers = [10, 5, 20, 4, 2, 10]
```

```python
#the variable a takes in turn each value of the list numbers
for a in numbers:
    print(a)
```

```python
print(a)
```

```python
animals = ['lion', 'cat', 'fish', 'ant']

for i in animals:
    print(i)
```

```python
print(numbers)

for num in numbers:
    if num >= 15:
        print(num)
```

```python
for num in numbers:
    if num != 10:
        print(num)
```

```python
#we try to count the number of elements in the list
counter = 0 

for i in numbers:
    counter = counter + 1

print(counter)
```

```python
len(numbers)
```

```python
for i in range(0,5):
    print(i)
```

```python
range(5) # iterable from 0 to 5 with a step of 1
```

```python
range(2, 10) # iterable from 2 to 10 with a step of 1
```

```python
for i in range(2, 10):
    print(i)
```

```python
range(2, 10, 2) # iterable from 2 to 10 with a step of 2
```

```python
for i in range(2, 10, 2):
    print(i)
```

```python
# another type of loop
stoper = 0

while (stoper < 10):
    stoper = stoper + 1
    #print(stoper)
print(stoper)
```
