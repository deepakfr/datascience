## Loops and strings

```python
for i in "camille":
    print(i)
```

```python
name = "Camille"
for i in name:
    print(i)
```

```python
phrase = "Here is a longer sentence!"
for i in phrase:
    print(i)
```

```python
print(name[0])
print(name[1])
print(name[2])
print(name[3])
print(name[4])
print(name[5])
print(name[6])
```

```python
name = "Camille"
for i in range(len(name)):
    print(name[i])
```

```python
for i in range(0,len(name)):
  j = i + 1
  print(j)
```
