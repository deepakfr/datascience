## Naming your variables and functions

clean-code-expressive-variable-names-basmati-rice.png

To name your variables and functions correctly, they must be named after what they represent, their value or their purpose. Obviously avoid naming your variables and functions with the names reserved by python (like if, else, list, max, ...).

Can you guess what the function defined below does ?  

```python
def stuff(a, b, c):
  for i in b:
    if a == i:
      c += 1
  return c
```

```python
def count_coins(coins_list, specific_coin, counter):
  for coin in coins_list:
    if coin == specific_coin:
      counter += 1
  return counter
```

There are also naming conventions in python for your variables and functions:
* Names cannot start with a number
* Names must be written **only** in lowercase with words separated by underscores
* Allowed characters are alphanumeric (numbers or letters) + underscore. Characters with accents are not allowed.
* Do not use the characters "l" (lower case letter L), "O" (upper case letter o) or "I" (upper case letter i) as single character variable names. In some fonts, these characters are indistinguishable from the numbers one and zero. 
* The preferred language is English.

## Spaces

The following operators are always surrounded by a space on both sides: 
* assignment (=)
* augmented assignment (+=, =, etc.)
* comparisons (==, <, >, !=, <>, <=, >=, in, not in, is, is not)
* booleans (and, or, not).

```python
#Correct 
a = 1
b = 2
if a > b and a != b:
  a += 1

#Wrong
a=1
b =2
if a> b and a!=b:
  a+= 1
```

Similarly, spaces are not added in the following situations:
* Immediately inside parentheses, brackets or braces
* Immediately before a comma, semicolon or colon
* Immediately before the open parenthesis that begins the list of arguments of a function call
* Immediately before the open parenthesis that starts an indexing or slicing

```python
# Correct:
spam(ham[1], {eggs: 2})

# Wrong:
spam( ham[ 1 ], { eggs: 2 } )
```

```python
# Correct:
if x == 4:
  print x, y
  x, y = y, x

# Wrong:
if x == 4 : 
  print x , y 
  x , y = y , x
```

```python
# Correct:
spam(1)

# Wrong:
spam (1)
```

```python
# Correct:
dct['key'] = lst[index]

# Wrong:
dct ['key'] = lst [index]
```

## Docstrings
Docstrings are special types of comments that appear right after the definition of functions or at the very beginning of our program. 
They can be accessed specifically to be read, unlike comments. Docstrings are always defined by a pair of 3 quotes: `'''` or `"""`.  For more details: https://www.datacamp.com/community/tutorials/docstrings-python

You can write function docstrings in one line:

```python
def say_hello(name):
    '''Takes in a name, returns a clean personnalized hello message'''
    message = "Hello " + name + " !"
    return message
```

```python
help(say_hello)
```

You can also write multi-line function docstrings:

```python
def say_hello(name):

    """Formats a clean hello message for a user

    Parameters
    ----------
    name : str
        The name of the user

    Returns
    -------
    str
        The hello message for the user
    """

    message = "Hello " + name + " !"
    
    return message
```
